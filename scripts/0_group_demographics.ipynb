## Load Master Demographic Dataframe

## Plot demographics by group

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(master_demo_df, x='DEMO_age', hue='Group', kde=True, ax=ax[0], bins=20, legend=False, palette='Set2', element='step')
# sns.histplot(master_demo_df, x='DEMO_age', hue='Group', kde=True, ax=ax[0], bins=20, legend=False, palette=['grey'], fill=False)
sns.boxplot(data=master_demo_df, x='Group', y='DEMO_age', ax=ax[1], palette='Set2', linewidth=2)
sns.stripplot(data=master_demo_df, x='Group', y='DEMO_age', ax=ax[1], color='grey', alpha=0.5, size=10)
ax[0].set_xlabel('')
ax[1].set_xlabel('')
ax[1].set_ylabel('Age')
sns.despine()
plt.show()

#ANOVA for Alcohol and Cannabis only
stat, p = stats.f_oneway(
    master_demo_df[master_demo_df['Group']=='Alcohol']['DEMO_age'],
    master_demo_df[master_demo_df['Group']=='Cannabis']['DEMO_age'],
)
print(f'Statistic: {stat}')
print(f'Pvalue: {p}')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
sns.histplot(master_demo_df, x='Group', hue='DEMO_sex', multiple='dodge', ax=ax, bins=20, shrink=0.8, palette='Set2', legend=False)
# Plot again without fill, outline color is black
sns.histplot(master_demo_df, x='Group', hue='DEMO_sex', multiple='dodge', ax=ax, bins=20, shrink=0.8, fill=False, palette=['grey'], legend=False)
ax.set_xlabel('')
sns.despine()
plt.show()
# Run chi square test for DEMO_sex across groups
cross_tabulation = pd.crosstab(master_demo_df['Group'], master_demo_df['DEMO_sex'])
display(cross_tabulation)
stat, p, _, _ = stats.chi2_contingency(cross_tabulation)
print(f'Statistic: {stat}')
print(f'Pvalue: {p}')
```
